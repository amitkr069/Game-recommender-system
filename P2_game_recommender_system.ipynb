# Vectorization

```python
import nltk
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=1000,stop_words='english')
```

```python
# new_game0 = segments[0]
```

```python
#no need of stemming

#direct creating vectors

vectors = cv.fit_transform(new_game['tags']).toarray()
vectors
```

```python
cv.get_feature_names_out()
```

```python
#distance calculation

from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_similarity(vectors)
```

```python
similarity = cosine_similarity(vectors)
similarity
```

```python
def recommend(game):
  game_index = new_game[new_game['title'] == game].index[0]
  distances = similarity[game_index]
  games_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[2:12]

  for i in games_list:
    print(new_game.iloc[i[0]].title)
```
